### Producers and Consumers

Consider the following solution to the producer-consumer problem:

```algorithm
empty, full := true, false
```

```algorithm
process producer
    var a: 0 .. N - 1 → T = …
    var p: 0 .. N
    p := 0
    while p < N do
        ⟨await empty then empty := false⟩
        buf := a(p)
        full, p := true, p + 1
```

```algorithm
process consumer
    var b: 0 .. N - 1 → T
    var c: 0 .. N
    c := 0
    while c < N do
        ⟨await full then full := false⟩
        b(c) := buf
        empty, c := true, c + 1
```
